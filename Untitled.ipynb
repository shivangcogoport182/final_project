```python
import pandas as pd 
import numpy as np
```

```python
train = pd.read_csv("/Users/shivangdudani/Desktop/house-prices-advanced-regression-techniques/train.csv")
test = pd.read_csv("/Users/shivangdudani/Desktop/house-prices-advanced-regression-techniques/test.csv")
train.head()
```

```python
target = train['SalePrice']
test_ids = test['Id']
train1 = train.drop(['Id','SalePrice'], axis=1)
test1 = test.drop('Id', axis=1)
data1 = pd.concat([train1, test1], axis=0).reset_index(drop=True)
data1
```

```python
data1.select_dtypes(object).isna().sum()
```

```python
def missing_percent(df,ty):
    nan_percent= 100*(df.select_dtypes(ty).isnull().sum()/len(df))
    nan_percent= nan_percent[nan_percent>0].sort_values()
    return nan_percent
```

```python
missing_percent(data1,np.number)
```

```python
data1['FireplaceQu'].unique()
```

```python
for column in [
    'Alley',
    'BsmtQual',
    'BsmtCond',
    'BsmtExposure',
    'BsmtFinType1',
    'BsmtFinType2',
    'FireplaceQu',
    'GarageType',
    'GarageFinish',
    'GarageQual',
    'GarageCond',
    'PoolQC',
    'Fence',
    'MiscFeature'
]:
    data1[column] = data1[column].fillna('Na')
  # Impute using the column mode
for column in [
    'MSZoning',
    'Utilities',
    'Exterior1st',
    'Exterior2nd',
    'MasVnrType',
    'Electrical',
    'KitchenQual',
    'Functional',
    'SaleType'
]:
    data1[column] = data1[column].fillna(data1[column].mode()[0])
```

```python
for column in data1.select_dtypes(np.number).isna().columns:
    data1[column] = data1[column].fillna(np.mean(data1[column]))
```

```python
data2 = data1.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
for column in data2.select_dtypes(object).columns:
    label = LabelEncoder()
    data2[column] = label.fit_transform(data2[column])
```

```python
data2
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
pca = PCA().fit(StandardScaler().fit_transform(data2))

plt.rcParams["figure.figsize"] = (20,6)

fig, ax = plt.subplots()
xi = np.arange(1, data2.shape[1]+1, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 79, step=1)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')

plt.show()
```

```python
pca1 = PCA(n_components = 60)
data3 = pd.DataFrame(pca1.fit_transform(data2),columns = ["PCA" + str(i) for i in range(1,61)])
```
